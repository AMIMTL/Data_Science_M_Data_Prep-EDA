```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import copy
computers = pd.read_csv("../Data/Computers.csv")
original_computers = copy.deepcopy(computers)

sns.set_style('darkgrid')
computers.tail()
```

### EDA

```python
sns.pairplot(computers , corner = True);
```

```python
computers.corr(numeric_only=True)
```

```python
(sns.heatmap(computers.corr(numeric_only=True) ,
             vmin = -1 ,
             vmax = +1 ,
             cmap = 'coolwarm' ,
             annot = True)
);
```

### First Model

```python
cols = ["ram", "speed", "hd", "screen"]
X = sm.add_constant(computers[cols])
y = computers['price']

model = sm.OLS(y, X).fit()

model.summary()
```

```python
sns.scatterplot(x = model.predict() , y = model.resid);
```

### Second Model

```python
cols = ["ram", "speed", "hd", "screen", "ads", "trend"]
X = sm.add_constant(computers[cols])
y = computers['price']

model = sm.OLS(y,X).fit()

model.summary()
```

```python
figure = sns.scatterplot(x = model.predict() , y = model.resid)
figure;
```

## Assignment 2: Metrics

Calculate MSE and RMSE for your regression model fitted on all numeric features.

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as rmse
from sklearn.metrics import r2_score as r2
X = sm.add_constant(computers[cols])
y = computers['price']

model = sm.OLS(y,X).fit()
print(f"MAE: {mae(y , model.predict(X))}")
print(f"RMSE squared: {rmse(y , model.predict(X))}")
print(f"RMSE: {rmse(y , model.predict(X),squared = False )}")
print(f"r2: {r2(y , model.predict(X))}")
```
